```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

pd.options.display.max_rows = 999
pd.options.display.max_columns = 999
```

```python
df = pd.read_csv('../data/nfl_play_by_play_2009_2018.csv')
df.head()
```

```python
# All columns after the 156th column, and in the cell below are either columns that will not 
# help in predicting, or the columns will support data leakage
df = df.iloc[:,:156]
```

```python
df.drop(columns=['posteam', 'posteam_type', 'defteam','game_date','time','yrdln','desc',
                 'qb_dropback','qb_kneel','qb_spike','qb_scramble','pass_length',
                 'pass_location','air_yards','yards_after_catch','run_location','run_gap',
                 'field_goal_result','kick_distance','extra_point_result',
                 'two_point_conv_result','timeout','timeout_team','td_team',
                 'total_home_score','total_away_score','posteam_score','defteam_score',
                 'score_differential','posteam_score_post','defteam_score_post',
                 'score_differential_post','punt_blocked','first_down_rush','first_down_pass',
                 'incomplete_pass','interception','punt_inside_twenty','punt_in_endzone',
                 'punt_out_of_bounds','punt_downed','punt_fair_catch','kickoff_inside_twenty',
                 'kickoff_in_endzone','kickoff_out_of_bounds','kickoff_downed',
                 'kickoff_fair_catch','own_kickoff_recovery','own_kickoff_recovery_td',
                 'qb_hit','rush_attempt','pass_attempt','sack','pass_touchdown',
                 'rush_touchdown','return_touchdown','extra_point_attempt','two_point_attempt',
                 'field_goal_attempt','kickoff_attempt','punt_attempt','complete_pass'], inplace=True)
```

```python
df.drop(columns=['play_id','game_id'], inplace=True)
```

```python
df.head()
```

I need to do some mapping. 
- All string objects that were null were changed to 'none'
- All null floats were changed to 0
- The home team, away team, and side of field columns were changed to dummies to help with feature engineering
I chose to make the string nulls to 'none' because dropping these columns would result in over 10,000 lines being dropped. I have a little over 100,000 that I will be working with and 10,000 is too many to drop. The same goes for the null floats. The models should pick up on this and adjust accordingly.

```python
df = df.apply(lambda x: x.fillna('none') if x.dtype == 'O' else x)
df = df.apply(lambda x: x.fillna(0.0) if x.dtype == 'float64' else x)
df = pd.get_dummies(data=df, columns=['home_team', 'away_team', 'side_of_field'], drop_first=True)
```

I transformed the target variables into integers
1. pass
2. run
3. kickoff
4. punt
5. field goal

```python
df.play_type = df.play_type.map(lambda x: 1 if x == 'pass' else x)
df.play_type = df.play_type.map(lambda x: 2 if x == 'run' else x)
df.play_type = df.play_type.map(lambda x: 3 if x == 'kickoff' else x)
df.play_type = df.play_type.map(lambda x: 4 if x == 'punt' else x)
df.play_type = df.play_type.map(lambda x: 5 if x == 'field_goal' else x)
```

Which half the game is currently in may help determine if the team wants to hold onto the ball or let it fly. I mapped the first half to == 1 and the second half to == 2.

```python
df.game_half = df.game_half.map(lambda x: 1 if x == 'Half1' else 2)
```

These were other play types. However, these plays have no direct impact on betting at a high level. These can be taken into account later. For now, they will be removed.

```python
df = df[~(df.play_type == 'extra_point')]
df = df[~(df.play_type == 'none')]
df = df[~(df.play_type == 'no_play')]
df = df[~(df.play_type == 'qb_kneel')]
df = df[~(df.play_type == 'qb_spike')]
```

```python
df = df.drop(columns=['side_of_field_none'])
```
